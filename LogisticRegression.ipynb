```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url = 'https://raw.githubusercontent.com/Schwarzam/Spiral-and-Eliptical-/master/group_2_zoo2.csv?token=APDNULOWV2BV2IVN2V464OC6WVPAU'
df = pd.read_csv(url)
```

```
ret = list (df)[0:16]
dem = list (df)[16:-1]
```

```
SC = df['simple_class']
```

```
def regular (cols):
  SC = cols[0]

  if SC == 'S':
    return 'S'
  elif SC == 'SB':
    return 'S'
  elif SC == 'Se':
    return 'S'
  else: 
    return 'E'

```

```
Nov = df[['simple_class']].apply(regular, axis=1)
```

```
Class = pd.get_dummies(Nov, drop_first=True)
```

```
df['Class'] = Class
df.drop(['simple_class'], axis=1, inplace=True)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[dem], df['Class'], test_size=0.3)
```

```
from sklearn.linear_model import LogisticRegression 
logis = LogisticRegression(max_iter=10000)
```

```
logis.fit(X_train, y_train)
```
